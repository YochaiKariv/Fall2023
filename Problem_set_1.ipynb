```python
# MNIST Loading and Visualization

from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

(X_train, y_train), (X_test, y_test) = mnist.load_data()
plt.figure(figsize=(4, 4))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(X_train[i], cmap='gray')
plt.show()
```

```python
# Random model implimantation

import numpy as np

# Flatten the images
X_train_flat = X_train.reshape(-1, 28 * 28)
X_test_flat = X_test.reshape(-1, 28 * 28)

# Randomly initialized matrix
random_matrix = np.random.rand(10, 28 * 28)

# y = Mx
y_pred = np.argmax(X_test_flat @ random_matrix.T, axis=1)

# Calculate Accuracy
accuracy = np.mean(y_pred == y_test)
print(f'Random Model Accuracy: {accuracy * 100:.2f}%')
```

```python
#Train Random Walk Model

best_acc = 0.0

for epoch in range(50):
    step = 0.01 * np.random.randn(10, 28 * 28)
    random_matrix += step
    y_pred_train = np.argmax(X_train_flat @ random_matrix.T, axis=1)
    acc = np.mean(y_pred_train == y_train)

    if acc > best_acc:
        best_acc = acc

    if best_acc > 0.75:
        print(f'Accuracy Exceeded 75%, stopping...')
        break
print(f'Random Walk Model Accuracy: {best_acc * 100:.2f}%')
```
