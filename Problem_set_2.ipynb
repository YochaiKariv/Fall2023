```python
from PIL import Image
import requests
from io import BytesIO

url = 'https://i.natgeofe.com/k/8fa25ea4-6409-47fb-b3cc-4af8e0dc9616/red-eyed-tree-frog-on-leaves-3-2_3x2.jpg'
response = requests.get(url)
img = Image.open(BytesIO(response.content))
img.show()
print(f"Shape of the image: {img.size}")
```

```python
img_resized = img.resize((224, 224))
img_resized.show()
```

```python
img_gray = img_resized.convert('L')
img_gray.show()
print(f"Shape of the grayscale image: {img_gray.size}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

img_array = np.array(img_gray)

for i in range(10):
    random_filter = np.random.rand(5, 5)
    convolved = convolve2d(img_array, random_filter, 'same')

    plt.figure()
    plt.subplot(1, 2, 1)
    plt.title('Random Filter')
    plt.imshow(random_filter, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title('Feature Map')
    plt.imshow(convolved, cmap='gray')

    plt.show()
```
